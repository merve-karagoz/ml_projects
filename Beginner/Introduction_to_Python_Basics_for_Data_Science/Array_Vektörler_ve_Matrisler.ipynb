**Array, Vektörler ve Matrisler**

**Array(diziler)**

*   Arrayler ve Python'un bir kombinasyonu size çok zaman kazandırabilir. Arrayler, kodunuzun genel boyutunu azaltmaya yardımcı olur.

**Array nedir?**

*   Bir listeye benzer şekilde, bir array aynı anda birden fazla değeri tutabilen bir veri yapısıdır.
*   Bununla birlikte, bir array yalnızca aynı veri türündeki sıralı öğeler koleksiyonunu tutabilir.











**Array ve Liste Örnekleri**


*   [1, 2, 3, 4, 5] bir integer arrayidir.

*   ['a', 'b', 'c', 'd', 'e'] bir string arrayidir.

*   [[1,2,3,4],
    [5,6,7,8]] 2 boyutlu bir arraydir.




**Listeler - Örnekler** 


*   [ 1, 2, 3, 'a', 5 ] hem tam sayıları hem de bir diziyi içeren bir listedir.


*   [ 1, 2, 3, 4, 'b', [ 1, 2, 3 ] ]









**Vektörler**

* Örneğin, [1, 2, 3, 4, 5] bir vektördür.
